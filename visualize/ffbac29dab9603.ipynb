```python
df = pd.read_csv('/kaggle/input/fish-market/Fish.csv')
df.head()
```

```python

df.info()
```

**No missing data**

```python
#handling species column categorical data
```

```python
df['Species'].unique()
```

```python
cat_df = pd.get_dummies(df['Species'], drop_first=True)
```

```python
df=pd.concat([df, cat_df], axis=1)
```

```python
df.head()
```

```python
df.drop(['Species'], axis=1 ,inplace=True)
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Weight', axis=1)
y=df.Weight
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
reg=model.fit(X_train, y_train)
```

```python
pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```

```python
from sklearn.model_selection import cross_val_score
corss_val = cross_val_score(reg, X_train, y_train, cv=10, scoring='r2')
```

```python
print(corss_val.mean())
```
